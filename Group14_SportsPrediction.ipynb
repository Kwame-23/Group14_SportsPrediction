**Training Models**

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest =train_test_split(x,y,test_size=0.2, random_state=42)
```

Checking which models out of Random Forest, GradientBoosting, AdaBoost with Desicion Tree as a base estimator

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
```

```python
rf=RandomForestRegressor(n_estimators=50,n_jobs=1)
```

```python
rf.fit(xtrain,ytrain)
```

```python
ypred = rf.predict(xtest)
```

Check for the mean absolute error (MAE) of the model

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(ypred,ytest)
```

Feature importance, checking the best feature that affect the accuracy of the model

```python
for name, score in zip(x.columns[:-1], rf.feature_importances_):
  print(name, " : ", score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb = GradientBoostingRegressor(n_estimators=50, learning_rate=0.1)
```

```python
gb.fit(xtrain, ytrain)
```

Check for the mean absolute error (MAE) of the model

```python
ypred = gb.predict(xtest)
mean_absolute_error(ypred,ytest)

```

Feature importance, checking the best feature that affect the accuracy of the model

```python
for name, score in zip(x.columns[:-1], gb.feature_importances_):
  print(name, " : ", score)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
abr = AdaBoostRegressor(base_estimator = dt, n_estimators=50, learning_rate=0.1)
```

```python
abr.fit(xtrain, ytrain)
```

```python
ypred=abr.predict(xtest)
```

Check for the mean absolute error (MAE) of the model

```python
mean_absolute_error(ypred,ytest) # adaboost gave the lower therefore it is the best model
```

Feature importance, checking the best feature that affect the accuracy of the model

```python
for name, score in zip(x.columns[:-1], abr.feature_importances_):
  print(name, " : ", score)
```

Cross Valiadation and Fine Turning AdaBoosting Parameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
c = KFold(n_splits=3)
params = {
    "n_estimators" : [50,100,300,500],
    "learning_rate" : [0.01, 0.1,0.2,0.3]
}
ab = AdaBoostRegressor(estimator = dt)

model_ab = GridSearchCV(ab,param_grid = params, cv =c)


model_ab.fit(xtrain,ytrain)
```

Check for best parameters

```python
model_ab.best_estimator_
```

```python
model_ab.best_params_
```

```python
y_pred=model_ab.predict(xtest)
```

```python
mean_absolute_error(y_pred, ytest)
```

Using the Optimised Parameters

```python
abr = AdaBoostRegressor(base_estimator = dt, learning_rate = 0.1, n_estimators = 300)
```

```python
abr.fit(xtrain, ytrain)
```

Saving the model using pickle


```python
with open('model.pkl','wb') as f:
  pickle.dump(abr,f)
```

```python
ypred=abr.predict(xtest)
```

```python
mean_absolute_error(ypred,ytest)
```
